```python
def load_data():
    # Cargar datos
    train_data_ext = pd.read_csv(ruta_dataset + 'train/train.csv')
    breed_labels = pd.read_csv(ruta_dataset+'breed_labels.csv')
    color_labels = pd.read_csv(ruta_dataset+'color_labels.csv')
    state_labels = pd.read_csv(ruta_dataset+'state_labels.csv')
    
    # Unir datos con etiquetas
    train_data_ext = train_data_ext.merge(breed_labels, left_on='Breed1', right_on='BreedID', how='left', suffixes=('', '_PrimaryBreed'))
    train_data_ext = train_data_ext.merge(breed_labels, left_on='Breed2', right_on='BreedID', how='left', suffixes=('', '_SecondaryBreed'))
    train_data_ext = train_data_ext.merge(color_labels, left_on='Color1', right_on='ColorID', how='left', suffixes=('', '_Color1'))
    train_data_ext = train_data_ext.merge(color_labels, left_on='Color2', right_on='ColorID', how='left', suffixes=('', '_Color2'))
    train_data_ext = train_data_ext.merge(color_labels, left_on='Color3', right_on='ColorID', how='left', suffixes=('', '_Color3'))
    train_data_ext = train_data_ext.merge(state_labels, left_on='State', right_on='StateID', how='left')
    
    return train_data_ext

# Llamar a la función y mostrar los datos unidos
train_data_ext = load_data()
print(train_data_ext.head())
```

```python
# Verificación de categorías únicas y su distribución
categorical_columns = ['Type', 'Breed1', 'Breed2', 'Gender', 'Color1', 'Color2', 'Color3', 'MaturitySize', 'FurLength', 'Vaccinated', 'Dewormed', 'Sterilized', 'Health', 'State']

for column in categorical_columns:
    print(f"Análisis de la columna: {column}")
    print(train_data[column].value_counts())
    print("\n")
```

```python
# Análisis de estadísticas descriptivas para variables numéricas
numerical_columns = ['Age', 'Quantity', 'Fee', 'VideoAmt', 'PhotoAmt', 'AdoptionSpeed']

for column in numerical_columns:
    print(f"Análisis de la columna: {column}")
    print(train_data[column].describe())
    print("\n")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Creación de boxplots para detectar valores atípicos
plt.figure(figsize=(15, 10))

for i, column in enumerate(numerical_columns, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(data=train_data, x=column)
    plt.title(f"Boxplot de {column}")

plt.tight_layout()
plt.show()
```

```python
# Distribución de la variable AdoptionSpeed
plt.figure(figsize=(10, 6))
sns.countplot(data=train_data, x='AdoptionSpeed')
plt.title('Distribución de la Velocidad de Adopción')
plt.xlabel('AdoptionSpeed')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Matriz de correlación
correlation_matrix = train_data[numerical_columns].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```
